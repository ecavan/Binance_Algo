```python
####GET Profit Margin from algo trading####

stocks = ["XRPUSDT", "TRXUSDT", "BNBUSDT", "ETHUSDT", "LTCUSDT", "BTCUSDT"]

def get_margin(stock):
    if (len(buy_signals[stock]) > 0) & (len(sell_signals[stock]) > 0):
        PM = (np.mean(sell_signals[stock]) - np.mean(buy_signals[stock]))*abs((quantity[stock]*len(sell_signals[stock]) - quantity[stock]*len(buy_signals[stock])))
        
    elif (len(buy_signals[stock]) > 0 )& (len(sell_signals[stock]) == 0):
        PM = (float(prices_inital[stock]) - np.mean(buy_signals[stock]))*(quantity[stock]*len(buy_signals[stock]))
                                                                      
    elif (len(buy_signals[stock]) == 0 )& (len(sell_signals[stock]) > 0):
        PM = (float(prices_inital[stock]) - np.mean(sell_signals[stock]))*abs((quantity[stock]*len(sell_signals[stock])))
    
    else:
        PM = 0
    
    return PM
                                                                      
for stock in stocks:
    print('Profit Margin ' + str(stock) + ' ' + str(get_margin(stock)))
```
